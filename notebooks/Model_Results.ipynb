```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# === Cargar mÃ©tricas
df = pd.read_csv("../models/model_performance.csv")

# Eliminar columnas no numÃ©ricas (como diccionarios de hiperparÃ¡metros)
df = df.drop(columns=["best_params"], errors="ignore")

# Confirmar que 'model' es de tipo str
df = df[df["model"].apply(lambda x: isinstance(x, str))]

# Reorganizar a formato largo
df_long = df.melt(id_vars="model", var_name="Metric", value_name="Score")

# === Graficar
plt.figure(figsize=(10, 6))
sns.barplot(data=df_long, x="Metric", y="Score", hue="model", palette="Set2")
plt.title("ðŸ” Model Performance Comparison")
plt.ylabel("Score")
plt.ylim(0, 1.05)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title="Model")
plt.tight_layout()
plt.savefig("../plots/evaluation/model_comparison.png")  # OPCIONAL
plt.show()

```
