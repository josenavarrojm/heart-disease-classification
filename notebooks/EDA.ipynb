```python
# Heart Disease Dataset - Exploratory Data Analysis
# Author: Jose Navarro Meneses

# üì¶ Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# üìê Set visual style
sns.set(style="whitegrid")
%matplotlib inline

# üìÅ Load dataset
df = pd.read_csv("../data/heart.csv")
print("‚úÖ Dataset loaded with shape:", df.shape)

# üîç First 5 rows
df.head()
```

```python
# üìä 1. Target Variable Distribution

sns.countplot(x="HeartDisease", data=df, palette="Set2")
plt.title("Distribution of Target Variable (HeartDisease)")
plt.xlabel("HeartDisease (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
# üìà 2. Basic statistics

df.describe(include="all").T
```

```python
# üß† 3. Distribution of key numerical features by class

num_cols = ["Age", "RestingBP", "Cholesterol", "MaxHR", "Oldpeak"]

for col in num_cols:
    plt.figure(figsize=(7, 4))
    sns.histplot(data=df, x=col, hue="HeartDisease", kde=True, bins=30, palette="Set1")
    plt.title(f"{col} Distribution by HeartDisease")
    plt.show()
```

```python
# üì¶ 4. Boxplots by target

for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x="HeartDisease", y=col, data=df, palette="Set2")
    plt.title(f"{col} vs HeartDisease")
    plt.show()
```

```python
# üßÆ 5. Categorical feature analysis

cat_cols = ["Sex", "ChestPainType", "RestingECG", "ExerciseAngina", "ST_Slope"]

for col in cat_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=col, hue="HeartDisease", data=df, palette="Set3")
    plt.title(f"{col} by HeartDisease")
    plt.xticks(rotation=30)
    plt.show()
```

```python
# üî• 6. Correlation heatmap

plt.figure(figsize=(12, 10))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
